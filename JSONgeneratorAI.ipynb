1. Falcon-7b-instruct

```python
from langchain import HuggingFacePipeline
from transformers import AutoTokenizer, pipeline
import torch
```

```python

model = "tiiuae/falcon-7b-instruct"

tokenizer = AutoTokenizer.from_pretrained(model)

f_pipeline = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    torch_dtype=torch.bfloat16,
    trust_remote_code=True,
    device_map="auto",
    do_sample=True,
    top_k=10,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id
)
```

```python
llm = HuggingFacePipeline(pipeline = f_pipeline, model_kwargs = {'temperature':0.7})
```

```python
from langchain import PromptTemplate, LLMChain

template = """
Instruction:
Return JSON format with datatype for {form_name} in the following format-
Response:
"""
prompt = PromptTemplate(template=template, input_variables=["form_name"])
```

```python
llm_chain = LLMChain(prompt=prompt, llm=llm)

question = "school registration form field"

print(llm_chain.run(question))
```

8. MPT-7B-instruct

```python
import transformers

model = transformers.AutoModelForCausalLM.from_pretrained(
  'mosaicml/mpt-7b-instruct',
  trust_remote_code=True
)
```

```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("EleutherAI/gpt-neox-20b")

pipe = pipeline('text-generation', model=model, tokenizer=tokenizer, device='cuda:0')

with torch.autocast('cuda', dtype=torch.bfloat16):
    print(
        pipe('Here is a recipe for vegan banana bread:\n',
            max_new_tokens=100,
            do_sample=True,
            use_cache=True))
```
